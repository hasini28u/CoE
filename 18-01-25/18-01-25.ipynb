```python
#display all the records of jan-2016
uber_data['START_DATE*'] = pd.to_datetime(uber_data['START_DATE*'], errors='coerce')

df = uber_data[(uber_data['START_DATE*'].dt.month == 1) & (uber_data['START_DATE*'].dt.year == 2016)]

print(df)
```
