```python
import pandas as pd
df = pd.read_csv("GlobalLandTemperaturesByCity.csv")
df
```

```python
median_value = df['AverageTemperature'].median()
```

```python
df['AverageTemperature'] = df['AverageTemperature'].fillna(median_value)
```

```python
df
```

```python
median_uncertainty = df['AverageTemperatureUncertainty'].median()
df['AverageTemperatureUncertainty'] = df['AverageTemperatureUncertainty'].fillna(median_uncertainty)
```

```python
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
#x-axis : year
#y-axis : averageTemperature
df['dt'] = pd.to_datetime(df['dt'], format='%Y-%m-%d')
df['Year'] = df['dt'].dt.year
df
```

```python
yearly_avg_temp = df.groupby('Year')['AverageTemperature'].mean()
```

```python
plt.plot(yearly_avg_temp.index, yearly_avg_temp.values, color='b', linestyle='-', markersize=5)
plt.show()
```

```python
#x-axis : countries
#y-axis : avg. temperature
import matplotlib.pyplot as plt
country_avg_temp = df.groupby('Country')['AverageTemperature'].mean()
country_avg_temp = country_avg_temp.dropna()

top_countries = country_avg_temp.sort_values(ascending=False).head(10)
plt.figure(figsize=(20, 6))
plt.bar(top_countries.index, top_countries.values)
plt.title("Top 10 Countries by Average Temperature")
plt.xlabel("Countries")
plt.ylabel("Average Temperature")
plt.show()
```

```python
#x-axis : temperature ranges
#y-axis : frequency
#using histogram

df = df.dropna(subset=['AverageTemperature'])

plt.figure(figsize=(10, 6))
plt.hist(df['AverageTemperature'], bins=5, color='skyblue', edgecolor='black')

plt.title('Distribution of Average Temperatures')
plt.xlabel('Average Temperature (°C)')
plt.ylabel('Frequency')
plt.show()
```

```python
#x-axis : avg. temp
#y-axis : temp. uncertainty 
#scatter plot 
df['AverageTemperature'] = df['AverageTemperature'].dropna()
df['AverageTemperatureUncertainty'] = df['AverageTemperatureUncertainty'].dropna()
plt.scatter(df['AverageTemperature'],df['AverageTemperatureUncertainty'],color='skyblue',s=10,alpha=0.5)
plt.title("Scatter Plot")
plt.xlabel("AverageTemperature")
plt.ylabel("AverageTemperatureUncertainty")
plt.show()
```

```python
#boxplot: spread of temperatures for a specific country over a given period
#x-axis : years or countries
#y-axis : avg. temp
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.boxplot(data = df[(df['Country']=='Denmark') & (df['Year']>=2000)], x='Year', y='AverageTemperature')
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df_filtered = df[(df['Country'] == 'Denmark') & (df['Year'] >= 2000)]
grouped_data = [df_filtered[df_filtered['Year'] == year]['AverageTemperature'] for year in df_filtered['Year'].unique()]

plt.figure(figsize=(10, 6))
plt.boxplot(grouped_data, labels=df_filtered['Year'].unique())

plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.title('Spread of Average Temperatures in Denmark (2000 and later)')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(data=tips , x='total_bill',bins=20)
plt.title("Distribution of total bills")
plt.show()
```

```python
#kde:kernel density estimation
sns.histplot(data=tips , x='total_bill',bins=20,kde=True)
plt.title("Distribution of total bills")
plt.show()
```

```python
sns.barplot(data=tips, x='day', y='total_bill', hue= 'time' ,palette='viridis')
plt.title("Average Total Bill by Day")
plt.show()
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue="sex",palette='coolwarm')
plt.title("Box plot of total bills by day and gender")
plt.show()
```

```python
sns.lineplot(data=tips,x='size',y='tip',marker='o',color="green")
```

```python
sns.pairplot(data=tips,hue='sex',palette='husl')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame(tips)

df_encoded = pd.get_dummies(df, drop_first=True)

corr = df_encoded.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)

plt.figure(figsize=(10, 6))
plt.show()
```
