```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [2,4,6,8]
plt.plot(x,y)
plt.title("Line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.boxplot(x,y)
plt.show()
```

```python
plt.plot(x,y,marker='*',linestyle='dotted')
plt.show()
```

```python
months = [1,2,3,4,5,6]
savings = [2000,3000,9000,1000,500,5000]
plt.plot(months,savings,color='green',linestyle='dashed',marker='o')
plt.title("growth of a savings account over 6-months")
plt.xlabel("months")
plt.ylabel("savings")
plt.show()
```

```python
categories = ['A','B','C','D','E']
values = [10,15,7,12,9]
plt.bar(categories,values,color = 'skyblue')
plt.title("Basic Bar chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt
ages=[20,22,22,23,25,27,27,28,30,31,32,34,35,35]
b = [20,30,40,50,60,70]
plt.hist(ages,bins=b,color='skyblue',edgecolor = 'black')
plt.title("Age Distribution of Customers")
plt.xlabel("Age Ranges")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

languages = ['C', 'Python', 'Java', 'C++']
percentages = [2.0, 30.0, 50.0, 18.0]
colors = ['skyblue', 'pink', 'yellow', 'violet']  # Corrected color list
plt.pie(percentages, labels=languages, autopct='%1.1f%%', startangle=180, colors=colors)
plt.legend()
plt.title('Programming Language Usage')
plt.show()
```

```python
x = [5,10,15,20,25]
y = [10,20,25,30,35]
plt.scatter(x,y,color='green',s=100,edgecolor='black')
plt.title("Scatter Plot Example")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
