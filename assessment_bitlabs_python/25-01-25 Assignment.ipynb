```python
import pandas as pd
df=pd.read_csv("supermarket_sales - Sheet1.csv")
df
```

```python
import matplotlib.pyplot as plt
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce', format='%m/%d/%Y')

df['YearMonth'] = df['Date'].dt.to_period('M')
```

```python
monthly_sales = df.groupby('YearMonth')['Total'].sum()
```

```python
# Use line plots track monthly sales and histograms for order quantity distribution
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', marker='o', color='b', linestyle='-', linewidth=2, markersize=5)
plt.title('Monthly Sales Over Time', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.show()
```

```python
# Plot Order Quantity Distribution (Histogram)
plt.hist(df['Quantity'], bins=20, color='green', edgecolor='black')
plt.title('Order Quantity Distribution', fontsize=14)
plt.xlabel('Order Quantity', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
data = pd.read_csv("Billionaires Statistics Dataset.csv")
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Age Distribution Histogram
sns.histplot(data['age'], kde=True, bins=20, color='skyblue')
plt.title('Age Distribution of the Individuals')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Suspending Patterns (Age vs Final Worth Scatter Plot)
sns.scatterplot(x=data['age'], y=data['finalWorth'], color='purple')
plt.title('Age vs Final Worth')
plt.xlabel('Age')
plt.ylabel('Final Worth (in millions)')
plt.show()
```

```python
data1 = pd.read_csv("egypt_education_dataset.csv")
data1
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

subjects = ['Subject_1', 'Subject_2', 'Subject_3', 'Subject_4', 'Subject_5', 'Subject_6', 'Subject_7', 'Subject_8', 'Subject_9', 'Subject_10']

student_data = data1.loc[0, subjects]

sns.barplot(x=subjects, y=student_data)
plt.title("Student's Marks Across Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
# Plotting box plot for all students' marks across subjects
plt.figure(figsize=(12, 8))
sns.boxplot(data=data1[subjects])
plt.title("Distribution of Marks Across Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```
